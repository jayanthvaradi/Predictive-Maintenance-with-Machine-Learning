```python
# Importing Required Libraries
import numpy as np                  # For numerical operations
import pandas as pd                 # For data manipulation and analysis
import matplotlib.pyplot as plt     # For plotting and visualization
import seaborn as sns               # For statistical data visualization
```

```python
# Reading a CSV File Using Pandas
df = pd.read_csv("C:\\Users\\acer\\Desktop\\Machine Learning\\Classification\\dataset2.csv") # `pd.read_csv()` reads the file and creates a structured DataFrame
```

```python
# Dropping Unnecessary Columns from the DataFrame
df = df.drop(columns = (["UDI","Product ID"]), axis = 1) # Drop the columns 'UDI' and 'Product ID' from the DataFrame
# `df.drop()` removes specified columns or rows from the DataFrame
# `columns=["UDI", "Product ID"]`: Specifies the columns to drop
# `axis=1`: Indicates that the operation is column-wise
```

```python
# Defining Input Variables
Input_variables = ["Air temperature [K]", "Process temperature [K]", "Rotational speed [rpm]", "Torque [Nm]", "Tool wear [min]"]
```

```python
# Filtering Data to Exclude Rows with "No Failure"
df = df[df["Failure Type"] != "No Failure"]
```

```python
# Displaying the DataFrame
print(df)
```

```python

# Creating Boxplots for Input Variables by Failure Type

# Create a 2x3 grid of subplots with a figure size of 15x6 inches
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(15, 6)) 

# Flatten axes for easier indexing
axes = axes.flatten()  

# Loop through input variables and plot each as a boxplot
for i, feature in enumerate(Input_variables):
    sns.boxplot(x=feature, y="Failure Type", data=df, ax=axes[i])

# Remove extra axes if fewer input variables than subplots
if len(Input_variables) < len(axes):
    fig.delaxes(axes[-1])

# Adjust layout for neat fitting
plt.tight_layout(rect=[0, 0, 1, 0.95])

# Display the plot
plt.show()
```
